## *Data preparation and cleaning.*

#### **Preparing and cleaning the telcom datasets**

```
#changing data column names for the three datasets 

t1.columns=['PRODUCT', 'VALUE','DATE_TIME','CELL_ON_SITE', 'DW_A_NUMBER','DW_B_NUMBER','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
t1.head()
```

```
t2.columns=['PRODUCT', 'VALUE','DATE_TIME','CELL_ON_SITE', 'DW_A_NUMBER','DW_B_NUMBER','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
t2.head()
```

```
t3.columns=['PRODUCT', 'VALUE','DATE_TIME','CELL_ON_SITE', 'DW_A_NUMBER','DW_B_NUMBER','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
t3.head()
```

```
#Merging the three tables into one.

data= [t1, t2, t3]
Telcom= pd.concat(data, axis=0, ignore_index=True)
Telcom.head()
```

```
# Dropping tables that won't be of help

Telcom.drop(['CELL_ON_SITE', 'COUNTRY_A', 'COUNTRY_B'], axis=1, inplace=True)
Telcom
```

#### Cleaning

```
#Getting duplicates

Telcom.duplicated()
```

```
#Dropping duplicates
Telcom_drop= Telcom.drop_duplicates()
Telcom_drop
```

```
#checking for null values
Telcom_drop.isnull()
```

```
# Dropping missing observations
Tel_drop= Telcom_drop.dropna()
Tel_drop
```

```
#dropping rows where all cells of a row is NA

Telco_drop=Tel_drop.dropna(how='all')
Telco_drop
```

```
#Filling in missing data with 0

Telcom_cleaned=Telco_drop.fillna(0)
Telcom_cleaned
```

#### Cleaning.

```
#checking for duplicates

# cells= cell
# cells

cellcol.duplicated()
```

```
#dropping duplicated values.

cell_drop=cellcol.drop_duplicates()
cell_drop
```

```
#checking for null values
cell_drop.isnull()
```

```
#dropping missing observations

celldropped=cell_drop.dropna()
celldropped
```

```
#Preparing the Telcom_cleaned for merging.

Telcom_cleaned = Telcom_cleaned.rename(columns={'SITE_ID': 'SITE_CODE'})
Telcom_cleaned
```

```
#merging the two dataframes

merged = pd.merge(left=Telcom_cleaned, right=cells_cleaned, left_on='SITE_CODE', right_on='SITE_CODE')
merged.shape
merged
```

```

merged.describe()
```

```
# separating date and time

merged['Dates'] = pd.to_datetime(merged['DATE_TIME']).dt.date
merged['Time'] = pd.to_datetime(merged['DATE_TIME']).dt.time
merged
```

```
# getting the total of billing price spent

grouped=merged.groupby(['PRODUCT', 'Dates'])['VALUE'].sum()
grouped
```
